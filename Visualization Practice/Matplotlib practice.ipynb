## Importing necessary libraries for plotting points using Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([0, 5])
y = np.array([0, 250])
```

## Plotting the points simply using a straight line

```python
plt.plot(x, y)
plt.show()
```

## Plotting the points using rings

```python
# 'o' indicates that we will now represent the mentioned points using rings (circles)
plt.plot(x, y, 'o')
plt.show()
```

## Plotting multiple points

```python
# pLotting random x and y coordinates
x_new = np.array([1, 2, 6, 8])
y_new = np.array([3, 8, 1, 10])

plt.plot(x_new, y_new)
plt.show()
```

If we do not specify the points on the x-axis, they will get the default values 0, 1, 2, 3 (etc., depending on the length of the y-points.

```python
y_new = np.array([3, 8, 1, 10, 5, 7])

plt.plot(y_new)
plt.show()
```

Using markers to emphasize each points we plot. There are various types of markers that can be used to design the points according the user's need.

```python
y_new = np.array([2, 4, 1, 9, 5, 10])

plt.plot(y_new, marker='o')
plt.show()
```

## Format Strings fmt
You can use also use the shortcut string notation parameter to specify the marker.

This parameter is also called fmt, and is written with this syntax:

###### marker|line|color

```python
y_new = np.array([2, 4, 1, 9, 5, 10])

# Here we've selected that marker would be 'o', line would be dotted :, and the colour would be green(g)
# Therefore it is - 'o:g'
plt.plot(y_new, 'o:g')
plt.show()
```

## Other marker features

Plotting with different marker size, using parameter - ms

```python
y_new = np.array([5, 1, 7, 3, 9, 2])

# We've set the marker size to 10
plt.plot(y_new, marker='o', ms=10)
plt.show()
```

###### Marker colour
You can select marker colour in two ways: 1. markeredgecolour (mec) and 2. markerfacecolour (mfc)

```python
y_points = np.array([2, 7, 1, 6, 9, 0, 4])

# Selecting marker edge colour as red
plt.plot(y_points, marker='o', ms=20, mec='r')
plt.show()

# Selecting marker face colour as red
plt.plot(y_points, marker='o', ms=20, mfc='r')
plt.show()

# Selecting both marker edge and face colour as red
plt.plot(y_points, marker='o', ms=20, mec='r', mfc='r')
plt.show()
```

## Matplotlib Lines

 We have a huge variety of features with which we can style the lines used in the plotting

```python
y = np.array([2, 7, 9, 3, 7, 4, 8])

# To specify that we are going to style the line, we use the parameter linestyle or ls
plt.plot(y, linestyle='dotted')
plt.show()

# We can also specify the colour of the line we want to plot
plt.plot(y, color='r')
plt.show()

# We can also change the width of the line that we are using to plot using the parameter linewidth or lw
plt.plot(y, linewidth=15)
plt.show()

# Combining all the features altogether
plt.plot(y, ls='dashed', color='g', lw=8)
plt.show()
```

## Multiple Lines
We can alo plot multiple lines on the same graph as follows:

```python
y1 = np.array([1, 6, 3, 8, 0, 3, 6, 8])
y2 = np.array([12, 4, 1, 18, 2 ,4, 1, 3])

# Plotting 2 curves together
plt.plot(y1)
plt.plot(y2)

plt.show()
```

## Matplotlib Labels

By using labels, we can give names to the x-axis and y-axis, and set a title for out graph too...
Along with that we can decorate our texts as well

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 6, 3.5, 7, 4])

plt.plot(x, y)

# Labelling the x and y axis
plt.xlabel("years")
plt.ylabel("sales in crore")

# Setting the title for the graph
plt.title("Record of sales of last 5 years")

plt.show()
```

###### Now, let us decorate the labels a bit
We can add font to our labels and change the position of title i.e., left, right or in the centre. By default the title position is in centre

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 6, 3.5, 7, 4])

font1 = {'family':'serif','color':'hotpink','size':16}
font2 = {'family':'serif','color':'darkred','size':12}

plt.plot(x, y)

# Labelling the x and y axis
plt.xlabel("years", fontdict=font2)
plt.ylabel("sales in crore", fontdict=font2)

# Setting the font for the title, and specifying its location
plt.title("Record of sales of last 5 years", fontdict=font1, loc='left')


plt.show()
```

## Matplotlib Grid
Now with plotting the points we can also set up the grid in the background

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 6, 3.5, 7, 4])

# Setting up the default grid
plt.plot(x, y)
plt.grid()
plt.show()

# Setting up the grid for x-axis only
plt.plot(x, y)
plt.grid(axis = 'x')
plt.show()

# Setting up the grid for y-axis only
plt.plot(x, y)
plt.grid(axis = 'y')
plt.show()

# Decorating the grid in the plotting graph
plt.plot(x, y)
plt.grid(color='green', ls='dashed', lw=0.7)
plt.show()
```

## Matplotlib Subplots
Now, by subplots we mean, creating multiple graphs within a single plot, adnd each such graph plotted in there are called subplots.

###### the subplot() function
The subplot function intakes 3 arguements namely the no. of rows, no. of columns and the current plot number respectively

```python
# plot 1
x = np.array([1, 5, 7, 3, 8, 0])
y = np.array([5, 7, 2, 6 ,8, 1])

# This subplot indicates that we will have 1 row, 3 columns, and this is the 1st plot
plt.subplot(1, 3, 1)
plt.plot(x, y)
plt.title("PLOT 1")

# plot 2
x = np.array([2, 9, 1, 8, 5, 3])
y = np.array([9, 4, 8, 2, 6, 1])

# This subplot indicates that we will have 1 row, 3 columns, and this is the 2nd plot
plt.subplot(1, 3, 2)
plt.plot(x, y)
plt.title("PLOT 2")

# plot 3
x = np.array([1, 9, 2, 8, 3, 7])
y = np.array([10, 20, 30, 40, 50, 60])

# This subplot indicates that we will have 1 row, 3 columns, and this is the 3rd plot
plt.subplot(1, 3, 3)
plt.plot(x, y)
plt.title("PLOT 3")

plt.show()
```

```python
# Now, the same thing will be displayed in multiple rows

# plot 1
x = np.array([1, 5, 7, 3, 8, 0])
y = np.array([5, 7, 2, 6 ,8, 1])

# This subplot indicates that we will have 3 rows, 1 columns, and this is the 1st plot
plt.subplot(3, 1, 1)
plt.plot(x, y)
plt.title("PLOT 1")

# plot 2
x = np.array([2, 9, 1, 8, 5, 3])
y = np.array([9, 4, 8, 2, 6, 1])

# This subplot indicates that we will have 3 rows, 1 columns, and this is the 2nd plot
plt.subplot(3, 1, 2)
plt.plot(x, y)
plt.title("PLOT 2")

# plot 3
x = np.array([1, 9, 2, 8, 3, 7])
y = np.array([10, 20, 30, 40, 50, 60])

# This subplot indicates that we will have 3 rows, 1 columns, and this is the 3rd plot
plt.subplot(3, 1, 3)
plt.plot(x, y)
plt.title("PLOT 3")

plt.show()
```

You can also give a super title for the main plot as follows:

```python
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x,y)
plt.title("SALES")

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.title("INCOME")

# We have set it as the super title which will appear on very top of the plot
plt.suptitle("MY SHOP")
plt.show()
```

## Matplotlib Scatter
Now instead of plotting the lines we will display the points in terms of dots are would be scattered rather than connected as a line

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

# plotting the points in terms of dots
plt.scatter(x, y)
plt.show()
```

Plotting and comparing more than two datasets on the same graph

```python
# Dataset 1
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y)

# Dataset 2
x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y)

plt.show()
```

```python
# If you want to specify the colours of each dataset, you can do it using color/c attribute
# Dataset 1
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y, color='green')

# Dataset 2
x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y, c='red')

plt.show()
```

```python
# Now, if you want to colour each dot int he plot, 
# you can put colours in an array and assign their values to the color attribute

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array(["red","green","blue","yellow","pink","black","orange","purple","beige","brown","gray","cyan","magenta"])

plt.scatter(x, y, c=colors)

plt.show()
```

```python
# The Matplotlib module has a number of available colormaps.
# A colormap is like a list of colors, where each color has a value that ranges from 0 to 100.
# Starting from 0 - Violet till 100 - Yellow
# This colour map is called virdis and you can include this colourmap in your plot using colormap/cmap attribute

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

# Setting the colour ranges
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

# Setting the coolourmap using cmap attribute
plt.scatter(x, y, c=colors, cmap='viridis')

plt.show()
```

You can also adjust the sizes of each dot on the scatter plot using the size(s) attribute

```python
# Creating dataset
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x, y, s=sizes)
plt.show()
```

Along with the sizes of the dots, we can also adjust the transparancy of the dots using alpha attribute

```python
# Creating dataset
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x, y, s=sizes, alpha=0.5)
plt.show()
```

Let's make a beautiful scatter plot by combining colour, size and transparancy of the dots

```python
# Creating random dataset
x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))

# Setting random colours to those data points
colors = np.random.randint(100, size=(100))

# Making each random size ten times the original value
sizes = 10 * np.random.randint(100, size=(100))

# We have used nipy_spectral color map to colour our dots
plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral')

plt.show()
```

## Matplotlib Bars
The bar() funtion is used to create bar graphs using pyplot in Matplotlib

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([10, 5, 7, 11])

plt.bar(x, y)
plt.show()

# We can also represent this bar graph horizontally using barh() function
plt.barh(x, y)
plt.show()
```

```python
# Along with the orientation, we can also change colours of the bar graphs

x = np.array(["A", "B", "C", "D"])
y = np.array([10, 5, 7, 11])
plt.subplot(1, 3, 1)
plt.bar(x, y, color='red')

x = np.array(["A", "B", "C", "D"])
y = np.array([6, 1, 8, 9])
plt.subplot(1, 3, 2)
plt.bar(x, y, color='green')

x = np.array(["A", "B", "C", "D"])
y = np.array([12, 23, 9, 16])
plt.subplot(1, 3, 3)
plt.barh(x, y, color='pink')

plt.show()
```

Now we can also set the height and width of the bars we are using in the bar graph.
###### Default bar width and height is 0.8

```python
# We can adjust the width of the bars in the bar graph using width attribute
x = np.array(["A", "B", "C", "D"])
y = np.array([10, 5, 7, 11])

# We can adjust the width of the bars when we are plotting them vertically
plt.bar(x, y, width=0.3)
plt.show()

# We can adjust the height of the bars when we are plotting them horizontally
plt.barh(x, y, height=0.3)
plt.show()
```

## Matplotlib Pie Charts
We can reperesent array values as pie charts using the pie() function

```python
x = np.array([35, 25, 18, 10])

plt.pie(x)
plt.show()
```

Each part in this pie chart i.e. the each section is called a wedge

The size of each wedge is determined by comparing the value with all the other values, by using this formula:
The value divided by the sum of all values: x/sum(x)

###### Labels
Using the label feature we can add name for each wedge. Label parameter must be an array consisting name for each wedge in the chart

```python
x = np.array([35, 25, 18, 10])
label = np.array(["Apple", "Mango", "Banana", "Cherry"])

plt.pie(x, labels=label)
plt.show()
```

###### startangle()
By default the start angle for drawing the pie charts is 0, and it goes in counter clockwise direction
So, using this parameter, we can set the angle at which we want to start our chart to start

```python
x = np.array([35, 25, 18, 10])
label = np.array(["Apple", "Mango", "Banana", "Cherry"])

# You can set any angle you want
plt.pie(x, labels=label, startangle=90)
plt.show()
```

###### Explode and Shadows
Many times we want some of the wedge to stand out, maybe the biggest or the smallest one.

But by default all the wedges are connected and we cannot make anyone stand out, so to do so, we use the explode() parameter.
The explode parameter must be an array containing the values that how much explosion you want.

###### shadow() parameter
This is used to give a 3D effect to the pie chart and take values either True or False

```python
x = np.array([35, 25, 18, 10])
label = np.array(["Apple", "Mango", "Banana", "Cherry"])
myexplode = np.array([0.2, 0, 0, 0])

plt.pie(x, labels=label, explode=myexplode)
plt.show()

# Now just to give the pie chart a 3D effect, use the shadow() parameter.
# This parameter would be either True or False, depending on whether you want it or not
x = np.array([35, 25, 18, 10])
label = np.array(["Apple", "Mango", "Banana", "Cherry"])
myexplode = np.array([0.2, 0, 0, 0])

plt.pie(x, labels=label, explode=myexplode, shadow=True)
plt.show()
```

###### Legend and Headers
Legend gives the explaination for each wedge, like it goves out information about the name of the wedge, colour used to represent that wedge.

And headers are just the head title for the legend used in the pie chart reprsentation

```python
x = np.array([35, 25, 18, 10])
mylabel = np.array(["Apple", "Mango", "Banana", "Cherry"])

plt.pie(x, labels=mylabel)
# Representing the Legend for the pie chart
plt.legend()
plt.show()

plt.pie(x, labels=mylabel)
# Representing the Legend along with the header
plt.legend(title="Fruits")
plt.show()
```

###### Pie chart colours
We can decide that what wedge should be of which colour using colors parameter. It should be an array containing colours for each wedge in the pie chart

```python
x = np.array([35, 25, 18, 10])
label = np.array(["Apple", "Mango", "Banana", "Cherry"])
mycolours = np.array(["pink", "blue", "green", "black"])

# Setting up the colours for wedge using the defined array
plt.pie(x, labels=label, colors=mycolours, shadow=True)
plt.show()
```
